```python
# Quantos estabelecimentos de saúde existem na base de dados?
len(df)
```

```python
## Top 10 bairros com mais estabelecimentos de saúde
top_bairros = df["BAIRRO"].value_counts().head(10)
top_bairros
```

```python
#Quais CEPs possuem a maior quantidade de estabelecimentos de saude?

cep = (
    df["CEP"]
    .value_counts()
    .reset_index()
)

cep.columns = ["CEP", "Quantidade"]

cep.head(10)

```

```python
#Quantos estabelecimentos possuem CNPJ próprio informado?
df["CNPJ PROPRIO"].notna().sum()
```

```python
#Quantos estabelecimentos possuem CNPJ da mantenedora informado?
df["CNPJ MANTENEDORA"].notna().sum()
```

```python
# Quantos bairros diferentes possuem pelo menos um estabelecimento de saúde?
df["BAIRRO"].nunique()
```

```python
# Quais logradouros possuem mais estabelecimentos?
df["LOGRADOURO"].value_counts().head(10)
```

```python
# Quais razões sociais possuem mais unidades cadastradas?
df["RAZAO SOCIAL"].value_counts().head(10)
```

```python
#"Os dez bairros com maior número de estabelecimentos concentram 
#aproximadamente X% dos registros da base."
(
    df["BAIRRO"]
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
    .head(10)
)
```
